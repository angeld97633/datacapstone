## Cleaning Data

```python
data = data.drop(['Transaction_date'], axis=1)
data = data.drop(['Transaction_ID'], axis=1)
data = data.drop(['Marital_status'], axis=1)
data = data.drop(['State_names'], axis=1)
data = data.drop(['Segment'], axis=1)
data = data.drop(['Employees_status'], axis=1)
data = data.drop(['Payment_method'], axis=1)
data = data.drop(['Referal'], axis=1)
```
